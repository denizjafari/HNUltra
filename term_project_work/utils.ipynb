```python
from os import path
import os
```

```python
parent_dir = "mis_images"
if not path.exists(parent_dir):
    os.mkdir(parent_dir)
    
for s in lines:
    if s == '\n':
        continue
    else:
        s = s.replace('\n', '')
        parts = s.split('\t')
        folder_name = path.join(parent_dir, parts[1]+"_"+parts[2])
        file_path = path.join(folder_name, parts[0]+'.jpg')
        print(folder_name, file_path)
```
